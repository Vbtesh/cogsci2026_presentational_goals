```python
import os
import csv
from PIL import Image, ImageDraw
```

```python
graphics_dir = '../images/'
scenes_dir = 'scenes/'
exp_dir = '../../docs/images/'

canvas_size = (200, 220)
line_width = 2
r = 12

images = {
    'R': Image.open(os.path.join(graphics_dir, 'agent_red.png')),
    'G': Image.open(os.path.join(graphics_dir, 'agent_green.png')),
    'P': Image.open(os.path.join(graphics_dir, 'agent_purple.png')),
    'A': Image.open(os.path.join(graphics_dir, 'apple_green.png')),
    'O': Image.open(os.path.join(graphics_dir, 'orange.png'))
}

colors = {
    'white': (255, 255, 255),
    'black': (0, 0, 0)
}
```

```python
positions = {
    'sim': {
        'R': (23, 110),
        'G': (123, 110),
        'P': (73, 15),
        'AOAO': [(26, 175), (50, 175), (126, 175), (150, 175)],
        'R-A': {'AOAO': (38, 187)},
        'R-O': {'AOAO': (62, 187)},
        'G-A': {'AOAO': (138, 187)},
        'G-O': {'AOAO': (162, 187)}
    },
    'seq': {
        'R': (73, 110),
        'G': (51, 88),
        'P': (73, 15),
        'AO': [(76, 175), (100, 175)],
        'AAO': [(64, 175), (88, 175), (112, 175)],
        'AOO': [(64, 175), (88, 175), (112, 175)],
        'AAOO': [(52, 175), (76, 175), (100, 175), (124, 175)],
        'R-A2': {
            'AO': (88, 187),
            'AAO': (76, 187),
            'AOO': (76, 187),
            'AAOO': (64, 187)
        },
        'R-A': {
            'AO': (88, 187),
            'AAO': (100, 187),
            'AOO': (76, 187),
            'AAOO': (88, 187)
        },
        'R-O2': {
            'AO': (112, 187),
            'AAO': (124, 187),
            'AOO': (124, 187),
            'AAOO': (112, 187)
        },
        'R-O': {
            'AO': (112, 187),
            'AAO': (124, 187),
            'AOO': (100, 187),
            'AAOO': (136, 187)
        }
    }
}
```

```python
with open('scene_info.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
        canvas = Image.new('RGB', canvas_size, colors['white'])
        draw = ImageDraw.Draw(canvas)
        draw.rectangle([(0, 0), (canvas_size[0]-line_width/2, canvas_size[1]-line_width/2)],
               outline=colors['black'], width=line_width)
        filename = row['type']
        # agents
        canvas.paste(images['G'], positions[row['type']]['G'], mask=images['G'])
        canvas.paste(images['R'], positions[row['type']]['R'], mask=images['R'])
        if row['P'] == 'present':
            canvas.paste(images['P'], positions[row['type']]['P'], mask=images['P'])
        # objects
        for i, obj in enumerate(row['objects']):
            canvas.paste(images[obj], positions[row['type']][row['objects']][i], mask=images[obj])
        if row['type'] == 'seq':
            filename += '_' + row['objects']
        # choices
        draw.circle(positions[row['type']]['R-'+row['R_choice']][row['objects']],
                    r, outline=colors['black'], width=line_width)
        filename += '_R-' + row['R_choice']
        if row['G_choice'] != '':
            draw.circle(positions[row['type']]['G-'+row['G_choice']][row['objects']],
                        r, outline=colors['black'], width=line_width)
            filename += '_G-' + row['G_choice']
        if row['div'] != '':
            draw.line([(100, 95), (100, 205)], fill=colors['black'], width=line_width*2)
            # add divider
            filename += '_div'
        filename += '_P-' + row['P'] + '.png'
        canvas.save(os.path.join(scenes_dir, filename))
        canvas.save(os.path.join(exp_dir, filename))
```
